### Here are some types of analysis you can do:

- Draw histograms and histogram variants for each feature or column.  (Swarm plot, kde plot, violin plot).


```python

numeric_columns = df.drop(columns= ["Year", "PolicyNumber"],).select_dtypes(include=["number"]).columns
```

```python
for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30, color="orange", edgecolor="black")
    plt.title(f"Histogram of {col}")
```

```python
for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    
    numeric_data = pd.to_numeric(df[col], errors='coerce').dropna()
    
    if len(numeric_data.unique()) > 1:
        sns.kdeplot(numeric_data)
        plt.title(f"KDE Plot of {col}")
```

```python
for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    sns.violinplot(y=df[col])
    plt.title(f"Violin Plot of {col}")
```

```python
for col in numeric_columns:
    plt.figure(figsize=(8, 4))
    
    sampled_data = df[col].dropna().sample(n=min(500, len(df)), random_state=42)
    
    sns.swarmplot(x=[0] * len(sampled_data), y=sampled_data, size=3)
    plt.title(f"Swarm Plot of {col}")
    plt.xticks([])
```


- Draw grouped histograms.  For instance, if you have tree heights for both maple and oak trees, you could draw histograms for both.


```python
df_fraud = df[df["FraudFound"] == "Yes"]
df_fraud_grouped = df_fraud.groupby("Sex")
df_fraud_male = df_fraud_grouped.get_group("Male")
plt.figure(figsize=(10, 6))
sns.histplot(data=df_fraud_male, x="AgeOfPolicyHolder", bins=20, alpha=0.5, color="blue")
plt.title("The Age of Male Fraud")

```

```python
df_fraud = df[df["FraudFound"] == "Yes"]
df_fraud_grouped = df_fraud.groupby("Sex")
df_fraud_male = df_fraud_grouped.get_group("Female")
plt.figure(figsize=(10, 6))
sns.histplot(data=df_fraud_male, x="AgeOfPolicyHolder", bins=20, alpha=0.5, color="blue")
plt.title("The Age of Female Fraud")
```


- Draw a bar plot to indicate total counts of each categorical variable in a given column.


```python
categorizal_columns = df.select_dtypes(include=["object"]).columns
for col in categorizal_columns:
    category_counts = df[col].value_counts()
    plt.figure(figsize=(15, 10))
    sns.barplot(x=category_counts.index, y=category_counts.values)
    plt.title(f"Count of Each Category in {col}")
    plt.xlabel(col)
    plt.ylabel("Count")
    
```
